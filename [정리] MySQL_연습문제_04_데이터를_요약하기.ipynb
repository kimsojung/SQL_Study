COUNT는 해당 열에 (예를 들면 breed에) 몇개의 값이 있는지 출력합니다. COUNT를 사용하면 데이터를 모두 출력하지 않고 몇개의 행이 있는지 알아볼 수 있습니다. 

그리고 연습 문제 3에서 배운 DISTINCT절을 COUNT와 같이 사용하면 해당 열에 고유의 값이 몇 개가 있는지 출력할 수 있습니다. 하지만 DISTINCT절은 괄호 안에 넣어야 하는 것을 잊지 마세요. 예를 들면 다음의 쿼리를 실행하면 품종 (breed)열에 몇 개의 고유 값이 있는지 출력할 수 있습니다. 

```SQL
SELECT COUNT(DISTINCT breed)
  FROM dogs
```

만약에 몇 마리의 고유한 개가 테스트를 적어도 한번 완료했는지 알고 싶다면 어떻게 해야할까요?

complete_tests 테이블의 데이터는 테스트가 완료되야만 생성되기 때문에 complete_tests 테이블에 있는 개들은 (Dog-Guid) 적어도 하나의 테스트는 완료했다고 볼 수 있습니다. 그리고 열의 이름이 (예를 들면 Dog_guid) 괄호 안에 들어가 있으면 미확정 값 (NULL)은 제외되고, 세어지지 않습니다. 따라서 몇마리의 고유한 개가 테스트를 적어도 한번 완료했는지 알고 싶으면 다음 쿼리를 실행하면 됩니다:

```SQL
SELECT COUNT(DISTINCT Dog_Guid)
  FROM complete_tests
```

** 질문 1 : 위의 쿼리를 WHERE절과 함께 사용해서 2014년 3월 1일 이후에 테스트를 완료 한 개는 몇마리가 있는지 알아보세요 (정담은 13,289행입니다). **

````python
%sql SELECT COUNT(DISTINCT Dog_Guid) FROM complete_tests WHERE created_at > "2014-03-01";
````

** 질문 2 : 두 번째 차이점을 알려면 COUNT (\*)를 사용하여 dogs 테이블의 행수를 계산하십시오. 여기서 주의해야 할 점은 COUNT뒤에 공백이 없이 (\*)를 명시해야 합니다. **

````python
%sql SELECT COUNT(*) FROM dogs;
````

** 질문 3 : 이제 dogs 테이블의 제외 (exclude)열에 있는 행의 수를 계산하십시오. **

````python
%sql SELECT COUNT(exclude) FROM dogs;
````

COUNT (\*)와 COUNT(열 이름)의 차이  

1. DISTINCT는 COUNT (\*)와 함께 사용할 수 없음
2. COUNT(열 이름)은 NULL값을 제외하고 계산하나, COUNT (\*)은 NULL값을 포함함

** 질문 4 : dogs 테이블에 제외 (exclude)가 된 행이 몇 개가 있습니까? 답은 853입니다 (힌트: exclude가 된 값을 찾으려면 exclude열의 값을 잘 살펴보세요 )**

````python
%sql SELECT SUM(exclude) FROM dogs;
````

## 2. SUM 함수

첫번째 쿼리의 결과가 
```mySQL
SELECT COUNT(exclude) 
FROM dogs
```
두번째 쿼리의 결과보다 훨씬 적다는 것은 
```mySQL
SELECT COUNT(*)
FROM dogs
```

바로 exclude열에 미확정 값 (NULL)이 많다는 것을 의미. 

ISNULL(첫번째 열, 두번째 열)이면 첫번째 열의 값이 미확정 (NULL)이면 두번째 열의 값으로 출력하라는 의미

ISNULL은 미확정 값 (NULL)이 있는 행에 1을 반환하게 하고 미확정 값이 아닌 행에는 0을 반환하게 하는 논리 함수.ISNULL (exclude) 구문을 SUM 안에 넣으면 미확값이 총 몇 개가 있는지 출력.

```mySQL
SELECT SUM(ISNULL(exclude))
FROM dogs
```

** dogs테이블의 exclude열의 미확정 값 (NULL)이 몇개가 있는지 계산해보십시오 : **

````python
%sql SELECT SUM(ISNULL(exclude)) FROM dogs;
````

## 3. AVG, MIN과 MAX 함수

AVG (평균), MAX (최고값), MIN (최소값)은 SUM (합계)과 비슷하게 작용함. 

** 질문 5 : "Memory versus Pointing" 게임에 평균, 최소 및 최대 등급 (rating)은 무엇입니까? (답은 각각 3.5584, 0 및 9이어야합니다.) **

````python
%sql SELECT test_name, AVG(rating) AS AVG_rating, MIN(rating) AS MIN_rating, MAX(rating) AS MAX_rating FROM reviews WHERE test_name="Memory versus Pointing";
````

````python
%sql SELECT test_name, AVG(rating),MIN(rating),MAX(rating) FROM reviews GROUP BY test_name ; 
````
