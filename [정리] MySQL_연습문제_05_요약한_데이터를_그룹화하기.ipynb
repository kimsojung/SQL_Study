그룹은 파생한 새로운 열을 통해서도 만들수 있음 
```mySQL
SELECT test_name, MONTH(created_at) AS Month, COUNT(created_at) AS Num_Completed_Tests
FROM complete_tests
GROUP BY Month;
```

또한 동시에 여러개의 원래 열과 파생된 열로 그룹을 짤수도 있음

위의 쿼리를 처음에는 GROUP BY절에 test_name, MONTH 순서로 넣어서 실행하고, 두번째에는 GROUP BY절에 MONTH, test_name의 순서로 넣어서 실행

````python
%sql SELECT test_name, MONTH(created_at) AS Month, COUNT(created_at) AS Num_Completed_Tests FROM complete_tests GROUP BY test_name, MONTH(created_at);
````

````python
%sql SELECT test_name, MONTH(created_at) AS Month, COUNT(created_at) AS Num_Completed_Tests FROM complete_tests GROUP BY MONTH(created_at), test_name;
````

첫 번째 쿼리의 결과를 살펴보면 첫 번째 열인 test_name에 같은 테스트 이름이 나와 있지만 완료된 테스트 (Num_Completed_Tests)의 수는 달별로 그룹을 나눔. 

두 번째 쿼리의 결과를 보면 첫 번째 열인 test_name이 달의 순서대로 정렬되어 있고, 테스트 별로 그룹을 나눔. 이 쿼리 결과를 출력하면 아래와 같음. 

<img src="https://duke.box.com/shared/static/y6l3ldxg289brid5mmn886qweglt65qv.jpg" width=800 alt="ORDER" />

데이터베이스마다 출력된 결과를 특정 방식으로 정렬하지만 그룹화 된 쿼리 뒤에 ORDER BY를 사용하면 원하는데로 정렬할 수 있음. 

```mySQL
SELECT test_name, MONTH(created_at) AS Month, COUNT(created_at) AS Num_Completed_Tests
FROM complete_tests
GROUP BY test_name, Month
ORDER BY test_name ASC, Month ASC;
```

MySQL을 포함한 일부 데이터베이스 서버에서는 GROUP BY 또는 ORDER BY절에 열의 이름 대신 숫자를 사용해서 쿼리의 길이를 줄일 수 있음
```mySQL
SELECT test_name, MONTH(created_at) AS Month, COUNT(created_at) AS Num_Completed_Tests
```

test_name이 1번, Month가 2번, Num_Completed_Tests는 3번이 됨. 그래서 위의 쿼리를 다음과 같이 다시 작성할 수 있음.

```mySQL

FROM complete_tests
GROUP BY 1, 2
```

## HAVING절 

WHERE절과 마찬가지로 HAVING절을 사용하면 원하는 데이터를 추릴 수 있음. WHERE절은 데이터의 각 행을 추릴 수 있다면 HAVING절은 데이터를 그룹으로 나눠서 추릴 수 있음. 

```mySQL
SELECT test_name, MONTH(created_at) AS Month, COUNT(created_at) AS Num_Completed_Tests
FROM complete_tests
WHERE MONTH(created_at)=11 OR MONTH(created_at)=12
GROUP BY 1, 2
HAVING COUNT(created_at)>=20
ORDER BY 3 DESC;
```


** 질문 3 : 질문 2에서 작성한 쿼리를 수정하여 (1) 품종 그룹(breed_group)열에서 미확정 값 (NULL)과 빈 문자열 (None)을 제외 시키고 (2) 최소 1,000 개의 고유의 Dog_Guids가 없는 그룹을 제외 하세요. 쿼리 결과에는 8행이 있어야 합니다. 
(힌트 : 때로 빈 문자열은 미확정 값(NULL)이 아닌 값으로 등록됩니다. 이 값들을 제외하려면 쿼리에서 다음의 조건을 넣어야 합니다. **

```mySQL
breed_group!=""
```


````python
%sql SELECT gender,breed_group, COUNT(DISTINCT(dog_guid)) FROM dogs WHERE breed_group IS NOT NULL AND breed_group!="None" AND breed_group!="" GROUP BY 1,2 HAVING COUNT(dog_guid)>=1000 ORDER BY 1 DESC;
````

** 질문 4 : 품종 유형 (breed type), 그 유형마다 완성 된 테스트의 평균 수, 그 유형의 평균 테스트 간 간격 (mean inter-test-interval)을 출력해보세요. 품종 유형으로 데이터를 정렬하려면 GROUP BY에는 어떤 열을 명시해야 할까요? 완성 된 테스트의 평균 수의 내림차순으로 정렬하셔야 합니다. (답에는 popular hybrid가 제일 먼저 출력되어 있게 됩니다).**

** 질문 5 : exam_answers 테이블을 사용해서 고객들이 테스트를 완료하는데 걸리는 평균 시간을 구하되, 6000분 이상의 개별 반응 시간 (individual reaction times)은 제외하고 평균 값을 구하세요. 그리고 출력된 결과에서 평균 반응 시간이 0분 보다 큰 것만 추리세요. (답으로 67 행이 출력되야 합니다) **

````python
%sql SELECT test_name, AVG(TIMESTAMPDIFF(minute, start_time, end_time)) AS TIME FROM exam_answers WHERE TIMESTAMPDIFF(minute, start_time, end_time) < 6000 GROUP BY test_name HAVING TIME > 0; 
````

** 질문 7 : 6번 문제와 같이 미국의 주 (State)와 우편번호의 각 조합에 해당하는 사용자 ID (User_Guids)의 총 수를 출력하는 쿼리를 작성하세요. 전과 같이 고유의 사용자여야 합니다. 적어도 5명의 사용자가 있는 조합만 출력하십시오. 출력될 이 데이터는 먼저 주 (state)이름의 오름차순 (알파벳순)으로 정렬하고, 두번째로는 사용자 ID의 총 수의 내림차순으로 정렬하세요. (출력된 답의 첫 번째 주는 AZ이어야 하며 우편번호는 86303이여야 합니다) **

````python
%sql SELECT state, zip, COUNT(DISTINCT(user_guid)) AS count FROM users GROUP BY state,zip HAVING count > 5 ORDER BY 1 ASC,count DESC;
````
