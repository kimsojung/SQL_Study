## Left 와 Right Joins

Inner Join은 WHERE절에 다음과 같이 명시해서 테이블을 연결:

```mySQL
WHERE d.dog_guid=r.dog_guid
```

위와 같이 "="를 사용해서 테이블들을 연결하는 것은 "equijoin".
FROM다음에 JOIN이라고 명시하고 ON 뒤에 연결 방식을 표현 하는 것은 "traditional join"

```mySQL
SELECT d.dog_guid AS DogID, d.user_guid AS UserID, AVG(r.rating) AS AvgRating, COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type
FROM dogs d JOIN reviews r
  ON d.dog_guid=r.dog_guid AND d.user_guid=r.user_guid
GROUP BY d.user_guid
HAVING NumRatings > 9
ORDER BY AvgRating DESC
LIMIT 200
```

JOIN이라고 쓰면 MySQL은 INNER JOIN이라고 이해.

```mySQL
SELECT d.user_guid AS UserID, d.dog_guid AS DogID, 
       d.breed, d.breed_type, d.breed_group
FROM dogs d, complete_tests c
WHERE d.dog_guid=c.dog_guid AND test_name='Yawn Warm-up';
```

**질문 1: 위의 쿼리를 FROM/JOIN/ON구문을 사용해서 다시 써서 실행해보세요**

````python
%sql SELECT d.user_guid AS UserID, d.dog_guid AS DogID,d.breed, d.breed_type, d.breed_group\
FROM dogs d JOIN complete_tests c\
ON d.dog_guid=c.dog_guid WHERE test_name='Yawn Warm-up' LIMIT 100;
````


```mysql
SELECT r.dog_guid AS rDogID, d.dog_guid AS dDogID, r.user_guid AS rUserID, d.user_guid AS dUserID, AVG(r.rating) AS AvgRating, COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type
FROM reviews r LEFT JOIN dogs d
  ON r.dog_guid=d.dog_guid AND r.user_guid=d.user_guid
WHERE r.dog_guid IS NOT NULL
GROUP BY r.dog_guid
HAVING NumRatings >= 10
ORDER BY AvgRating DESC;
```

**질문 2: 위의 쿼리를 어떻게 수정하면 RIGHT JOIN을 써도 같은 정보를 출력할 수 있을까요?**

````python
%sql SELECT r.dog_guid AS rDogID, d.dog_guid AS dDogID, r.user_guid AS rUserID, d.user_guid AS dUserID, AVG(r.rating) AS AvgRating, COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type\
FROM reviews r LEFT JOIN dogs d\
ON r.dog_guid=d.dog_guid AND r.user_guid=d.user_guid\
WHERE r.dog_guid IS NOT NULL\
GROUP BY r.dog_guid\
HAVING NumRatings >= 10\
ORDER BY AvgRating DESC LIMIT 100;
````

````python
%sql SELECT r.dog_guid AS rDogID, d.dog_guid AS dDogID, r.user_guid AS rUserID, d.user_guid AS dUserID, AVG(r.rating) AS AvgRating, COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type\
FROM dogs d RIGHT JOIN reviews r\
ON r.dog_guid=d.dog_guid AND r.user_guid=d.user_guid\
WHERE r.dog_guid IS NOT NULL\
GROUP BY r.dog_guid\
HAVING NumRatings >= 10\
ORDER BY AvgRating DESC LIMIT 100;
````

JOIN 시 중복 ID가 존재하는 데이터가 key값으로 잡혀있을 경우, 중복된 수만큼 곱해져서 출력된다.  

그래서 테이블을 Join 할때는 다음 사항들을 참고해야함: 
+ 데이터를 분석할때는 테이블에 있는 행과 열이 고유 값이라고 믿지 말고 직접 결과를 출력해서 체크해보세요
+ 데이터를 집결 (i.e. 더하거나, 그룹을 하거나, 갯수를 세기)전에 항상 데이터를 출력해서 살펴보세요. Join을 할때는 특히 더 중요합니다.  
+ 테이블을 몇차례 Join하거나 집결할때는 각 Join과 집결 함수의 쿼리를 작은 단위로 미리 실행해보세요.  
+ 데이터와 그 분석결과를 맹신하지 마세요 -- 어떤 결과를 내리기 전에 데이터가 어떻게 출력되어 있는지 항상 체크해야 합니다.

Full outer join은 left join과 right join과 비슷하게 두 테이블의 모든 결과를 (NULL 값 포함) 다 출력함. 


<mark>**MySQL은 안타깝게도 Full outer join을 지원하지 않음** </mark>
   


## Outer Join¶

**질문 10: Left join을 사용하여 users 테이블에 있지만 dogs테이블에는 없는 고유의 user_guids의 갯수를 반환하는 쿼리는 어떻게 작성하겠습니까? (답은 2226행 입니다)**

````python
%sql SELECT COUNT(u.user_guid)\
FROM users u LEFT JOIN dogs d ON u.user_guid=d.user_guid\
WHERE d.user_guid IS NULL;
````

**질문 11: Right join을 사용하여 users 테이블에 있지만 dogs테이블에는 없는 고유의 user_guids의 갯수를 반환하는 쿼리는 어떻게 작성하겠습니까? (답은 2226행 입니다)**




````python
%sql SELECT COUNT(u.user_guid)\
FROM dogs d RIGHT JOIN users u ON u.user_guid=d.user_guid\
WHERE d.user_guid IS NULL;
````
