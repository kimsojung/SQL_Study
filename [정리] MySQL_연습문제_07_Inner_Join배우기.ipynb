## 1. 2개의 테이블을 연결하는 Inner Join

```python
%load_ext sql
%sql mysql://studentuser:studentpw@localhost/dognitiondb
%sql USE dognitiondb
```

```python
%sql SELECT user_guid, dog_guid, rating \
FROM reviews \
GROUP BY user_guid \
LIMIT 100;
```

왜 rating을 평균값과 갯수(count)로 집계하는지 이해하지 못했으나 이는 한 강아지의 review는 여러개가 존재했기 때문이었다. 

```python
%sql SELECT d.user_guid, d.dog_guid, d.breed_type, count(r.rating) AS count, AVG(r.rating) as AVG \
FROM dogs d, reviews r \
WHERE d.dog_guid=r.dog_guid AND d.user_guid=r.user_guid \
GROUP BY d.user_guid, d.dog_guid, d.breed_type \
HAVING count >= 10 \
ORDER BY AVG DESC \
LIMIT 200;
```

breed는 dog_guid와 동일 집계단위이므로 group by 문에서 실질적인 집계역할을 하지는 않으나, SELECT문에서 집계되지 않은 변수 명은 groupby의 절에 포함되어야한다는 원칙으로 인해 group by 뒤에 포함됨. 

```python
%sql SELECT d.dog_guid AS DogID, d.user_guid AS UserID, AVG(r.rating) AS AvgRating, \
COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type \
FROM dogs d, reviews r \
WHERE d.dog_guid=r.dog_guid AND d.user_guid=r.user_guid \
GROUP BY UserID, DogID, d.breed, d.breed_group, d.breed_type \
HAVING NumRatings >= 10 \
ORDER BY AvgRating DESC \
LIMIT 200;
```

앞 쿼리에서는 200명까지 출력하라고 했는데도 38개의 행인 테이블이 출력되었음, 그 이유는 10개의 이상의 리뷰를 남긴 개 주인의 수가 38명이기 때문임. 위의 쿼리에서 HAVING과 LIMIT절을 없에면 395 행의 테이블이 나옴

```python
%sql SELECT d.dog_guid AS DogID, d.user_guid AS UserID, AVG(r.rating) AS AvgRating, \
COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type \
FROM dogs d, reviews r \
WHERE d.dog_guid=r.dog_guid AND d.user_guid=r.user_guid \
GROUP BY UserID, DogID, d.breed, d.breed_group, d.breed_type \
ORDER BY AvgRating DESC;
```

위의 쿼리 결과를 통해 (A) 개 주인들은 리뷰들을 많이 남기지 않았다 (B) 리뷰를 남긴 결과를 살펴보면 결과에 그렇게 놀라지 않았다 는 것을 알수 있음

**밑에 있는 첫번째 셀에는 INNER JOIN의 쿼리를 실행하되 WHERE절에 dog_guid만 명시하고, 두번째 셀에는 user_guid만 명시해보세요:**

```python
%sql SELECT d.dog_guid AS DogID, d.user_guid AS UserID, AVG(r.rating) AS AvgRating, \
COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type \
FROM dogs d, reviews r \
WHERE d.dog_guid=r.dog_guid \
GROUP BY UserID, DogID, d.breed, d.breed_group, d.breed_type \
ORDER BY AvgRating DESC;
```

```python
%sql SELECT d.dog_guid AS DogID, d.user_guid AS UserID, AVG(r.rating) AS AvgRating, \
COUNT(r.rating) AS NumRatings, d.breed, d.breed_group, d.breed_type \
FROM dogs d, reviews r \
WHERE d.user_guid=r.user_guid \
GROUP BY UserID, DogID, d.breed, d.breed_group, d.breed_type \
ORDER BY AvgRating DESC;
```

dog_guid로만 테이블을 연결하면 395개의 행의 답이 출력되는 반면에 user_guid로만 테이블을 연결하면 6267개의 행의 답이 출력됨. 이 뜻은: 

+ reviews테이블에 있는 user_guid들은 모두 다 dogs 테이블에 있음
+ 6000개가 넘은 dog_guids가 reviews테이블에 있지만, 그중에서 395개의 dog_guids만 dogs테이블에 있음
